```python
# day  : last contact day of the month
# month: last contact month of year
# Drop 'month' and 'day' as they don't have any intrinsic meaning
df.drop('month', axis=1, inplace=True)
df.drop('day', axis=1, inplace=True)
df.drop('pdays', axis=1, inplace=True)
df.drop('pdays', axis=1, inplace=True)
```

```python
df['housing'] = df['housing'].map({'no':0,'yes':1})
df['housing']
```

```python
df['loan']= df['loan'].map({'yes':1,'no':0})
df['loan']
```

```python
df['y']=df['y'].map({'yes':1,'no':0})
df['y']
```

```python
#Now converting the values of columns in Numerical format in o's & 1's for Decsion tree 
df['default']= df['default'].map({'yes':1,'no':0})
df['default']
```

```python
# Convert categorical variables to dummies
df = pd.get_dummies(data=df, columns = ['job', 'marital', 'education', 'poutcome'], \
                                   prefix = ['job', 'marital', 'education', 'poutcome'])
df.head()
```

```python
df.columns
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier #2) [Used to create classifier]
from sklearn.tree import plot_tree
from sklearn.model_selection import train_test_split #1) [used tp train & split]
from sklearn.metrics import classification_report
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from matplotlib.colors import ListedColormap 
```

```python
x=[	'age', 'balance', 'housing', 'loan','poutcome_failure','poutcome_other','poutcome_success']
x=df[x]
y=df['y']
```

```python
print(x)
print(y)
```

```python
(x_train,x_test,y_train,y_test)=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
clf=DecisionTreeClassifier(max_depth=3)
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
plt.figure(figsize=(30, 16))
plot_tree(clf, filled=True, feature_names=x.columns.tolist(), rounded=True, fontsize=18)
plt.show()
```

```python
#Creating the Confusion matrix  
cm= confusion_matrix(y_test, y_pred) 
cm
#This indicates the 5+26= 31 incorrect predictions and 110+12=122 correct predictions.
```

```python
Accuracy = metrics.accuracy_score(y_test, y_pred)
Accuracy
```

```python
report = classification_report(y_test, y_pred)
print(report)
```
